```python
from skimage import data, color
import skimage.util as util
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = data.astronaut()
plt.imshow(image)
plt.show()
```

```python
image.shape
```

```python
plt.imshow(image[:,:,0], cmap='gray')
```

```python
astronauta_gray = color.rgb2gray(image)
plt.imshow(astronauta_gray, cmap='gray')
```

```python
vertical_flip_image = np.flipud(astronauta_gray)
plt.imshow(vertical_flip_image, cmap='gray')
```

```python
imagen_con_ruido = util.random_noise(astronauta_gray, mode='s&p') # s&p, gaussian
plt.imshow(imagen_con_ruido, cmap='gray')
```

```python
from skimage.filters import gaussian, median

imagen_suavizada = median(imagen_con_ruido, np.ones((3,3)))
plt.imshow(imagen_suavizada, cmap='gray')
```

```python
def comparar_imagenes():
    fig, axes = plt.subplots(1, 3, figsize=(8, 4))
    ax = axes.ravel()

    ax[0].imshow(astronauta_gray, cmap='gray')
    ax[0].set_title('Original image')

    ax[1].imshow(imagen_con_ruido, cmap='gray')
    ax[1].set_title('Image with noise')

    ax[2].imshow(imagen_suavizada, cmap='gray')
    ax[2].set_title('Smoothed image')

    for a in ax:
        a.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
comparar_imagenes()
```
